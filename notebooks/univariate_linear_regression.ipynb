```python
# Create the model
import importlib
import ml_algorithms.linear_regression
importlib.reload(ml_algorithms.linear_regression)

from ml_algorithms.linear_regression import UnivariateLinearRegression

# Prepare your dataset
X = df['YearsExperience'].values  # Years of Experience
y = df['Salary'].values           # Salary

# Create the model
model = UnivariateLinearRegression()

# Train the model with the real dataset
model.fit(X, y, learning_rate=0.01, epochs=1000)

model.print_parameters()

# Predict new values
# Predict salaries for new experience values
new_experience = [1.5, 3.0, 5.0]
predicted_salaries = model.predict(new_experience)

# Display the predictions in a clean and user-friendly format
print("Predicted Salaries based on Years of Experience:")
for exp, salary in zip(new_experience, predicted_salaries):
    print(f" - {exp} years: ₹{salary:,.2f}")


# Plot the fit line
model.plot_fit(X, y)
```
